###Importar arquivos para DBFS Landingzone

Para baixar um arquivo do GitHub e salvá-lo no Databricks, você pode seguir os passos abaixo:


```python

# URL do arquivo no GitHub
url = 'https://github.com/andrerosa77/trn-pyspark/raw/main/dados_2011.csv'

# Caminho de destino no Unity Catalog
# Formato: /Volumes/<catalog>/<schema>/<volume_name>/<path_within_volume>/<filename>
unity_catalog_path = '/Volumes/workspace/default/lhdw/landingzone/vendas/processar/dados_2011.csv'

# Baixar o arquivo diretamente do GitHub para o Unity Catalog usando dbutils.fs.cp
# O dbutils.fs.cp pode lidar com URLs HTTP/HTTPS como fonte.
dbutils.fs.cp(url, unity_catalog_path)

print(f"Arquivo baixado e salvo em: {unity_catalog_path}")

```
